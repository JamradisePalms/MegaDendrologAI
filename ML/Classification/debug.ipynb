```python
tree_labels = {}
has_hollow = {}
has_cracks = {}
has_fruits_or_flowers = {}
unique = set()
```

```python
import json
import os
import matplotlib.pyplot as plt
from PIL import Image

# --- Configuration ---
JSON_FILE = "/Users/pasheeee/MyProjects/MegaDendrologAI/ML/Classification/result/classification_result.json"
IMAGE_DIR = "/Users/pasheeee/MyProjects/MegaDendrologAI/ML/Classification/images"

def display_images_with_labels(json_path, images_base_dir, show=False):
    """
    Opens a JSON file, reads the data, and displays each image
    with its corresponding 'tree_type' as a title.
    """
    # 1. Open and read the JSON file
    with open(json_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # 2. Loop through each item in the data
    for item in data:
        label = item.get('tree_type', 'Unknown')
        hollow = item.get('has_hollow')
        crack = item.get('has_cracks')
        fruits = item.get('has_fruits_or_flowers')
        original_path = item.get('image')
        injury = item.get('injuries')

        tree_labels[label] = tree_labels.get(label, 0) + 1
        has_hollow[hollow] = has_hollow.get(hollow, 0) + 1
        has_cracks[crack] = has_cracks.get(crack, 0) + 1
        has_fruits_or_flowers[fruits] = has_fruits_or_flowers.get(fruits, 0) + 1

        injuries = injury.split(",")
        for inj in injuries:
            unique.add(inj)

        if show:
            if injury:
                if not original_path:
                    print(f"Skipping item with no image path: {item}")
                    continue

                filename = original_path

                img = Image.open(filename)

                plt.figure(figsize=(8, 8))
                plt.imshow(img)
                plt.title(f"Tree Type: {label}", fontsize=14)
                plt.axis('off')
                
                print(f"Showing image: {filename} | Label: {label}")
                plt.show()
```

```python
display_images_with_labels(JSON_FILE, IMAGE_DIR, show=False)
```
