```python
import json
import numpy as np
from collections import Counter
import torch

def calculate_class_weights(data, tasks):
    """
    Рассчитывает веса классов для многозадачной классификации
    Поддерживает как числовые, так и строковые метки
    """
    class_weights = {}
    
    for task_name, num_classes in tasks.items():
        print(f"\n=== {task_name} ({num_classes} классов) ===")
        
        # Собираем все метки для этой задачи
        labels = []
        for item in data:
            if task_name in item:
                labels.append(item[task_name])
        
        if not labels:
            print(f"  Предупреждение: нет данных для задачи '{task_name}'")
            continue
        
        # Подсчитываем количество примеров каждого класса
        class_counts = Counter(labels)
        total_samples = len(labels)
        
        print(f"  Всего примеров: {total_samples}")
        print(f"  Распределение классов: {dict(class_counts)}")
        
        # Создаем маппинг строковых меток в числовые индексы
        unique_labels = sorted(list(class_counts.keys()))
        label_to_index = {label: idx for idx, label in enumerate(unique_labels)}
        
        print(f"  Маппинг меток в индексы: {label_to_index}")
        
        # Рассчитываем веса (метод PyTorch style)
        weights_dict = {}
        for label, count in class_counts.items():
            weights_dict[label] = total_samples / (num_classes * count)
        
        # Создаем массив весов для PyTorch (в порядке индексов)
        weights_array = np.ones(num_classes)
        for label, weight in weights_dict.items():
            idx = label_to_index[label]
            if idx < num_classes:  # Проверка на выход за границы
                weights_array[idx] = weight
        
        class_weights[task_name] = {
            'counts': class_counts,
            'total_samples': total_samples,
            'label_to_index': label_to_index,
            'weights_dict': weights_dict,  # Веса по оригинальным меткам
            'weights_array': weights_array,  # Веса в виде массива для PyTorch
            'index_to_label': {v: k for k, v in label_to_index.items()}  # Обратный маппинг
        }
        
        # Выводим результаты
        print("  Веса классов:")
        for label in sorted(class_counts.keys()):
            count = class_counts[label]
            percentage = (count / total_samples) * 100
            weight = weights_dict[label]
            idx = label_to_index[label]
            print(f"    Класс '{label}' (индекс {idx}): {count} примеров ({percentage:.1f}%) -> вес: {weight:.2f}")
    
    return class_weights

def get_pytorch_weights_dict(class_weights):
    """
    Преобразует рассчитанные веса в формат для PyTorch
    """
    pytorch_weights = {}
    
    for task_name, task_data in class_weights.items():
        pytorch_weights[task_name] = {
            'weights': task_data['weights_array'],
            'label_mapping': task_data['label_to_index']
        }
    
    return pytorch_weights

# Основная часть выполнения
if __name__ == "__main__":
    # Загрузка данных
    with open("train_data.json", 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    # Задачи классификации
    tasks = {
        "tree_type": 27,
        "has_hollow": 2,
        "has_cracks": 2, 
        "has_fruits_or_flowers": 2,
        "overall_condition": 6,
        "has_crown_damage": 2,
        "has_trunk_damage": 2,
        "has_rot": 2
    }
    
    # Расчет весов
    class_weights = calculate_class_weights(data, tasks)
    
    # Получение весов в формате PyTorch
    pytorch_weights = get_pytorch_weights_dict(class_weights)
    
    print("\n" + "="*50)
    print("ВЕСА ДЛЯ PYTORCH:")
    print("="*50)
    
    for task_name, task_weights in pytorch_weights.items():
        print(f"\n{task_name}:")
        print(f"  Маппинг: {task_weights['label_mapping']}")
        print(f"  Веса: torch.tensor({list(task_weights['weights'])}, dtype=torch.float32)")
    
    print("\n" + "="*50)
    print("ПРИМЕР ИСПОЛЬЗОВАНИЯ:")
    print("="*50)
    
    for task_name in tasks.keys():
        if task_name in pytorch_weights:
            weights_tensor = f"torch.tensor({list(pytorch_weights[task_name]['weights'])}, dtype=torch.float32).to(device)"
            print(f"criterions['{task_name}'] = nn.CrossEntropyLoss(weight={weights_tensor})")
    
    print("\n" + "="*50)
    print("МАППИНГ МЕТОК ДЛЯ ПРЕОБРАЗОВАНИЯ:")
    print("="*50)
    
    for task_name, task_data in class_weights.items():
        print(f"\n{task_name}:")
        print(f"  {task_data['label_to_index']}")
```

```python
import os
import json
from collections import defaultdict

existing_json_path = r"C:\Users\shari\PycharmProjects\MegaDendrologAI\full_fixed_data_naming.json"
web_data_folder = r"C:\Users\shari\PycharmProjects\MegaDendrologAI\ML\web_data"
output_json_path = r"C:\Users\shari\PycharmProjects\MegaDendrologAI\combined_data.json"

def create_combined_json():
    with open(existing_json_path, 'r', encoding='utf-8') as f:
        existing_data = json.load(f)
    
    web_data_images = []
    unique_classes = set()
    for class_name in os.listdir(web_data_folder):
        class_path = os.path.join(web_data_folder, class_name)
        unique_classes.add(class_name)
        if os.path.isdir(class_path):
            for filename in os.listdir(class_path):
                if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                    full_image_path = os.path.join(class_path, filename)
                    
                    web_data_images.append({
                        "tree_type": class_name,
                        "image": full_image_path
                    })
    
    combined_data = []
    
    for item in existing_data:
        unique_classes.add(item["tree_type"])
        combined_data.append({
            "tree_type": item.get("tree_type", ""),
            "image": item.get("image", "")
        })
    
    combined_data.extend(web_data_images)
    
    with open(output_json_path, 'w', encoding='utf-8') as f:
        json.dump(combined_data, f, ensure_ascii=False, indent=2)
    print(len(unique_classes))
    print(f"Объединенный JSON создан: {output_json_path}")
    print(f"Всего записей: {len(combined_data)}")
    print(f"Из существующего JSON: {len(existing_data)}")
    print(f"Из web_data: {len(web_data_images)}")

create_combined_json()
```
