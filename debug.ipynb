```python
cnt = {}
cnt2 = {}
for item in data:
    injury = item['injuries'].split(',')
    for i in injury:
        if i not in cnt:
            cnt[i] = 0
        cnt[i] += 1

    if item['tree_type'] not in cnt2:
        cnt2[item['tree_type']] = 0
    cnt2[item['tree_type']] += 1
```

```python
import copy
cnt = 0
processed_data = copy.deepcopy(data)
for item in processed_data:
    new_injuries = []
    for injury in item['injuries'].split(','):
        if injury not in label_mapping:
            cnt += 1
            continue
        new_injuries.append(label_mapping[injury.strip()])
    
    item['injuries'] = ','.join(new_injuries)

with open('new_data.json', 'w') as f:
    json.dump(processed_data, f, indent=4)
```
